```python
import pandas as pd, numpy as np, matplotlib.pyplot as plt, seaborn as sb
```

```python
df = pd.read_csv('../data/new_movie.csv')
df
```

```python
X = df.loc[:, ['제작 비용 ($)', '상영 등급', '장르', '상영 시간', '개봉 연도']]
X
```

```python
y = df['영화 유형'].to_frame()
y
```

```python
from sklearn.preprocessing import LabelEncoder
from pandas.api.types import is_integer_dtype, is_float_dtype, is_object_dtype
```

```python
X_new = pd.DataFrame()
for col in X:
    if is_integer_dtype(X[col]) or is_float_dtype(X[col]): 
        X_new[col] = X[col]
    elif is_object_dtype(X[col]):
        encoder = LabelEncoder()
        X_new[col] = encoder.fit_transform(X[col])
```

```python
X_new
```

#### 데이터 분할

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_new, y, test_size = 0.2, random_state = 4)
```

#### KNN 통해 신규 데이터 분류

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
classifier = KNeighborsClassifier(n_neighbors = 5)
```

```python
classifier.fit(X_train, y_train)
```

#### y 계산

```python
y_pred = classifier.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
```

```python
cm = confusion_matrix(y_test, y_pred)
cm
```

```python
accuracy_score(y_test, y_pred)
```
