```python
import pandas as pd
```

```python
df_drink = pd.read_csv('../data/starbucks_drinkMenu_expanded.csv')
df_drink
```

```python
df_drink.columns = df_drink.columns.str.strip()
```

```python
df_drink.dtypes
```

```python
df_drink['Total Fat (g)'] = df_drink['Total Fat (g)'].str.replace(' ', '.')
df_drink['Vitamin A (% DV)'] = df_drink['Vitamin A (% DV)'].str.replace("%", '')
df_drink['Vitamin C (% DV)'] = df_drink['Vitamin C (% DV)'].str.replace("%", '')
df_drink['Calcium (% DV)'] = df_drink['Calcium (% DV)'].str.replace("%", '')
df_drink['Iron (% DV)'] = df_drink['Iron (% DV)'].str.replace("%", '')
```

```python
df_drink['Caffeine (mg)'] = df_drink['Caffeine (mg)'].str.replace('Varies', '-1')
df_drink['Caffeine (mg)'] = df_drink['Caffeine (mg)'].str.replace('varies', '-1')
```

```python
df_drink['Vitamin A (% DV)'] = df_drink['Vitamin A (% DV)'].astype(int)
df_drink['Vitamin C (% DV)'] = df_drink['Vitamin C (% DV)'].astype(int)
df_drink['Calcium (% DV)'] = df_drink['Calcium (% DV)'].astype(int)
df_drink['Iron (% DV)'] = df_drink['Iron (% DV)'].astype(float)
df_drink['Caffeine (mg)'] = df_drink['Caffeine (mg)'].astype(float)
```

```python
df_drink['Caffeine (mg)'] = df_drink['Caffeine (mg)'].replace(-1.0, df_drink['Caffeine (mg)'].mean().round())
```

```python
df_drink['Caffeine (mg)'] = df_drink['Caffeine (mg)'].fillna(df_drink['Caffeine (mg)'].mean())
```
